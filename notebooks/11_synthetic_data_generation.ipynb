```python
import sys
import numpy as np
# Add the project root directory to the Python path
sys.path.append('..')
from datetime import datetime
from src.synthetic_generation.synthetic_data_generator import SyntheticSensorDataGenerator

start_timestamp=datetime(2025, 1, 1, 8, 0)
end_timestamp=datetime(2025, 1, 1, 22, 0)

# Create generator for a specific time range
generator = SyntheticSensorDataGenerator(
    start_timestamp=start_timestamp,
    end_timestamp=end_timestamp
)

# Generate data
acc_df, hr_df = generator.generate_sensor_data_without_gaps()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from typing import Optional

    
    # Create figure with two subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 10), sharex=True)
    
    
acc_plot = acc_df
hr_plot = hr_df
    
    # Calculate accelerometer magnitude
acc_magnitude = np.sqrt(acc_plot['x']**2 + acc_plot['y']**2 + acc_plot['z']**2)
    
    # Plot accelerometer magnitude
ax1.plot(acc_plot['timestamp'], acc_magnitude, 'b-', label='Accelerometer Magnitude', alpha=0.7)
ax1.set_ylabel('Acceleration (m/s²)')
ax1.set_title('Accelerometer Magnitude')
ax1.grid(True)
    
    # Add gravity reference line
ax1.axhline(y=9.81, color='r', linestyle='--', alpha=0.5, label='Gravity (9.81 m/s²)')
ax1.legend()
    
    # Plot heart rate
    # Filter out invalid values for plotting
valid_hr = hr_plot[hr_plot['bpm'] != -9999]
ax2.plot(valid_hr['timestamp'], valid_hr['bpm'], 'g-', label='Heart Rate', alpha=0.7)
ax2.set_ylabel('Heart Rate (BPM)')
ax2.set_title('Heart Rate')
ax2.grid(True)
ax2.legend()
    
    # Format x-axis
plt.gcf().autofmt_xdate()  # Rotate and align the tick labels
ax2.set_xlabel('Time')
    
    # Add a title for the entire figure
plt.suptitle(f'Sensor Data from {start_timestamp.strftime("%Y-%m-%d %H:%M:%S")} '
                f'to {end_timestamp.strftime("%Y-%m-%d %H:%M:%S")}', y=1.02)
    
    # Adjust layout to prevent label cutoff
plt.tight_layout()
```

```python
# Print basic statistics
print("\nAccelerometer Data Statistics:")
print(f"Number of samples: {len(acc_df)}")
print(f"Time range: {acc_df['timestamp'].min()} to {acc_df['timestamp'].max()}")
print(f"Sampling rate: {len(acc_df) / ((end_timestamp - start_timestamp).total_seconds()):.1f} Hz")
print("\nAccelerometer magnitude statistics:")
acc_magnitudes = np.sqrt(acc_df['x']**2 + acc_df['y']**2 + acc_df['z']**2)
print(f"Mean magnitude: {acc_magnitudes.mean():.2f} m/s² ({acc_magnitudes.mean()/9.81:.3f} g)")
print(f"Std magnitude: {acc_magnitudes.std():.2f} m/s² ({acc_magnitudes.std()/9.81:.3f} g)")
    
print("\nHeart Rate Data Statistics:")
print(f"Number of samples: {len(hr_df)}")
print(f"Time range: {hr_df['timestamp'].min()} to {hr_df['timestamp'].max()}")
print(f"Sampling rate: {len(hr_df) / ((end_timestamp - start_timestamp).total_seconds()):.1f} Hz")
print(f"Mean heart rate: {hr_df['bpm'].mean():.1f} BPM")
print(f"Std heart rate: {hr_df['bpm'].std():.1f} BPM")
```

```python
# Save to OCED format
acc_df.to_json('data/synthetic/accelerometer_events.json', orient='records')
hr_df.to_json('data/synthetic/heartrate_events.json', orient='records')
```
